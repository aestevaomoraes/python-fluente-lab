## 💡 O mais importante

Uma **sequência** é uma coleção **ordenada** de elementos.

O aspecto mais importante não é que seus elementos estejam ordenados em ordem crescente ou alfabética, mas sim que **cada elemento ocupa uma posição conhecida (índice)**.

Exemplos de sequências:

```python
[10, 20, 30]

("SP", "RJ", "MS")

"Python"

range(10)
```

As sequências permitem operações comuns como:

- acesso por índice;
- iteração com `for`;
- slicing;
- busca de elementos;
- comparação entre sequências.

O Python divide as sequências em duas grandes categorias:

### Sequências Mutáveis

Podem ser modificadas após sua criação.

Exemplos:

- `list`
- `bytearray`
- `array.array`

### Sequências Imutáveis

Não podem ser alteradas após serem criadas.

Exemplos:

- `tuple`
- `str`
- `bytes`
- `range`

Embora possuam diferenças importantes, todas compartilham uma interface muito semelhante, permitindo que grande parte do código funcione da mesma forma independentemente do tipo utilizado.

## 🧠 Modelo Mental

Pense em uma sequência como uma estante de livros.

Cada livro ocupa uma posição específica.

```text
Índice

0 → Python
1 → SQL
2 → Pandas
3 → NumPy
```

O importante é que cada elemento possui uma posição previsível.

Isso permite ao Python acessar rapidamente qualquer elemento utilizando seu índice.
