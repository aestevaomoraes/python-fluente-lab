A REPL é o ambiente interativo do Python. Quando digitamos apenas o nome de um objeto, sem usar print(), ela utiliza o método especial __repr__() para decidir como esse objeto será exibido ao programador.

#### 🧠 Regra para Memorizar

• objeto (REPL ou pdb) → __repr__()

• print(objeto) → __str__()

• Se __str__() não existir → Python usa __repr__()

```python
f"{Produto!r}"
```

```python
repr (Produto)
```

🧠 Regra para Memorizar

O método especial `__repr__()` não é utilizado apenas pela REPL.

Ele também é chamado por:

- `repr(objeto)`
- `f"{objeto!r}"` (marcador de conversão `!r`)
- Ferramentas de depuração (debuggers)

Em todos esses casos, o objetivo é obter uma representação útil do objeto para o desenvolvedor.

🧠 Regra para Memorizar

Um bom `__repr__()` deve:

- Não gerar dúvidas sobre o objeto.
- Mostrar informações suficientes para identificar a classe e seus atributos.
- Sempre que possível, parecer um código Python capaz de recriar o próprio objeto.

#### 📝 Revisão

Nesta seção aprendemos que o Python utiliza os métodos especiais `__repr__()` e `__str__()` para controlar a forma como um objeto é representado.

- `__repr__()` fornece uma representação técnica, voltada para o desenvolvedor, sendo utilizada pela REPL, pela função `repr()` e pelo marcador `!r`.
- `__str__()` fornece uma representação amigável ao usuário e é utilizada pela função `print()`.
- Caso `__str__()` não seja implementado, o Python utiliza `__repr__()` como alternativa.
- Um bom `__repr__()` deve ser claro, não ambíguo e, sempre que possível, representar um código capaz de recriar o objeto.

Esses métodos tornam nossas classes mais integradas ao Python e facilitam a depuração, a leitura do código e o desenvolvimento de bibliotecas profissionais.

#### 💡 O mais importante

O Python não executa diretamente funções e operadores sobre os objetos. Em vez disso, ele faz perguntas aos próprios objetos por meio dos **métodos especiais (dunder methods)**.

| Código | O que o Python pergunta ao objeto? | Método Especial |
|--------|-------------------------------------|-----------------|
| `len(obj)` | Qual é o seu tamanho? | `__len__()` |
| `abs(obj)` | Qual é o seu valor absoluto? | `__abs__()` |

| `if obj:` | Você deve ser considerado verdadeiro ou falso? | *(Descobriremos nesta seção.)* |

📌 A grande ideia do Modelo de Dados do Python é que **o comportamento de um objeto é definido pelos métodos especiais implementados na sua classe**.

É isso que permite que objetos personalizados se comportem como tipos nativos da linguagem, tornando o código mais intuitivo, reutilizável e integrado ao ecossistema Python.

```python
class Carteira:

    def __init__(self, saldo):
        self.saldo = saldo

    def __bool__(self):
        return self.saldo > 0
```

```python
c1 = Carteira(100)
c2 = Carteira(0)

print(bool(c1))
print(bool(c2))
```

```python
class Carrinho:

    def __init__(self):
        self.itens = []

    def __len__(self):
        return len(self.itens)
```

```python
c = Carrinho()
print(bool(c))
```

```python
c.itens.append("Notebook")
print(bool(c))
```

Abstract Base Classes (ABCs) são classes que definem um contrato para outras classes.   
Elas indicam quais métodos ou propriedades uma classe filha deve implementar,     
sem necessariamente fornecer a implementação completa. Em Python, ficam no módulo abc.

```python
from abc import ABC, abstractmethod

class Forma(ABC):
    @abstractmethod
    def area(self):
        pass
```

```python
forma = Forma()  # TypeError
# Forma não pode ser instanciada diretamente:
```

```python
class Retangulo(Forma):
    def __init__(self, largura, altura):
        self.largura = largura
        self.altura = altura

    def area(self):
        return self.largura * self.altura
```

```python
retangulo = Retangulo(10, 5)
print(retangulo.area())  # 50
```

Até aqui aprendemos que:

*  Collection é qualquer objeto que armazena outros objetos.
*  API de Collections: define interfaces (contratos) que agrupam objetos com comportamentos semelhantes.
*  O Python organiza as coleções por comportamento, não apenas pelo tipo.

```python
# Uma tupla não é uma list, mas é uma sequência. Strings também:
from collections.abc import Sequence

dados = (10, 20, 30)
ndados = "10, 20, 30"
```

```python
isinstance(dados, list)      # False
isinstance(dados, Sequence)  # True
```

```python
isinstance(ndados, tuple)      # False
isinstance(ndados, Sequence) # True
```

```python
class Turma:

    def __init__(self):

        self.alunos = ["Ana","Carlos","Pedro"]
```

```python
class Turma:

    def __init__(self):

        self.alunos = ["Ana","Carlos","Pedro"]

    def __iter__(self):

        return iter(self.alunos)

    def __len__(self):

        return len(self.alunos)

    def __contains__(self, nome):

        return nome in self.alunos
```

```python
# O objeto ainda não existe. Você deveria criar uma instância.
# Isso funciona porque você está dizendo explicitamente qual objeto será usado como self.
turma = Turma()
```

```python
turma.__len__()
```

```python
len(turma)
```

```python
turma.__iter__()
```

```python
for aluno in turma:
    print(aluno)
```

```python
turma.__contains__("Ana")
```

```python
"Ana" in turma
```

Toda coleção Python deve responder a três perguntas:

📏 Quantos elementos existem?
→ __len__()

🔄 Como percorrer os elementos?
→ __iter__()

🔍 Um elemento pertence à coleção?
→ __contains__()

Use list quando seu código precisa de uma lista de verdade — por exemplo, quando vai usar .append() ou .sort() alterando o objeto.  
Use Sequence quando basta receber algo ordenado, com len(), índices e iteração, aceitando listas, tuplas, strings etc.  
Ou seja: list verifica a implementação concreta; Sequence verifica o contrato/comportamento esperado.

```python
class Livro:

    def __init__(self, titulo):
        self.titulo = titulo

    def __str__(self):
        return self.titulo

    def __len__(self):
        return len(self.titulo)

livro = Livro("Python Fluente")

print(livro)

print(len(livro))
```

```python
class Cidade:

    def __init__(self, nome):
        self.nome = nome

    def __str__(self):
        return self.nome

    def __len__(self):
        return len(self.nome)

    def __repr__(self):
        return f"Cidade(nome= {self.nome!r})"

cidade = Cidade("Campo Grande")

print(cidade)
print(repr(cidade))
```

```python
class Cidade:

    def __init__(self, nome):
        self.nome = nome

    def __str__(self):
        return self.nome


    def __repr__(self):
        return f"Cidade(nome= {self.nome!r})"

cidade = Cidade("Campo Grande")

print(cidade)
print(repr(cidade))
```

```python
texto = "Python"

print(texto)
print(str(texto))
print(repr(texto))
```

#### 📌 Regra de ouro (essa vale decorar)  

    * __str__ é a forma amigável para quem usa o programa.   

    * __repr__ é a forma detalhada para quem desenvolve o programa.   

Se um objeto estiver sozinho, normalmente vemos `__str__`.   
 
Se ele estiver dentro de outra estrutura (como uma lista, dicionário ou tupla), normalmente veremos `__repr__`.   

```python
class Cidade:

    def __str__(self):
        return "Campo Grande"

    def __repr__(self):
        return "Cidade('Campo Grande')"
```

```python
cidade = Cidade()

print(cidade)

print([cidade])

print((cidade,))
```

Uma lista pode conter dezenas de objetos diferentes.     
 Para representá-los de forma precisa e útil para o programador, ela utiliza repr() de cada elemento.
