## 📖 1.3.1 Emulando Tipos Numéricos

### 🎯 Objetivo

Nesta seção aprenderemos como fazer objetos personalizados se comportarem como tipos numéricos nativos do Python.

Por meio dos métodos especiais, podemos definir como nossos objetos respondem a operações como soma (`+`), valor absoluto (`abs()`), comparação e outras operações matemáticas.
```python
class Vetor:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __abs__(self):
        return (self.x**2 + self.y**2) ** 0.5
```

## 📖 1.3.2 Representação como String

### Objetivo

Nesta seção aprenderemos como personalizar a forma como um objeto é exibido no Python.

Por meio dos métodos especiais de representação, podemos definir como uma instância da nossa classe aparece quando é impressa no terminal, exibida no Jupyter Notebook ou inspecionada durante o desenvolvimento.

Esses métodos tornam os objetos mais legíveis, facilitam a depuração do código e melhoram a experiência de uso de bibliotecas como NumPy, pandas e outras ferramentas de Data Science.

```python
class Produto:
    def __init__(self, nome):
        self.nome = nome
```

```python
p = Produto('Notebook')
print(p)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Produto": ["Notebook", "Mouse"],
    "Preço": [3500, 80]
})

df
```

A REPL é o ambiente interativo do Python. Quando digitamos apenas o nome de um objeto, sem usar print(), ela utiliza o método especial __repr__() para decidir como esse objeto será exibido ao programador.

#### 🧠 Regra para Memorizar

• objeto (REPL ou pdb) → __repr__()

• print(objeto) → __str__()

• Se __str__() não existir → Python usa __repr__()

```python
f"{Produto!r}"
```

```python
repr (Produto)
```

🧠 Regra para Memorizar

O método especial `__repr__()` não é utilizado apenas pela REPL.

Ele também é chamado por:

- `repr(objeto)`
- `f"{objeto!r}"` (marcador de conversão `!r`)
- Ferramentas de depuração (debuggers)

Em todos esses casos, o objetivo é obter uma representação útil do objeto para o desenvolvedor.

🧠 Regra para Memorizar

Um bom `__repr__()` deve:

- Não gerar dúvidas sobre o objeto.
- Mostrar informações suficientes para identificar a classe e seus atributos.
- Sempre que possível, parecer um código Python capaz de recriar o próprio objeto.

## 📝 Revisão

Nesta seção aprendemos que o Python utiliza os métodos especiais `__repr__()` e `__str__()` para controlar a forma como um objeto é representado.

- `__repr__()` fornece uma representação técnica, voltada para o desenvolvedor, sendo utilizada pela REPL, pela função `repr()` e pelo marcador `!r`.
- `__str__()` fornece uma representação amigável ao usuário e é utilizada pela função `print()`.
- Caso `__str__()` não seja implementado, o Python utiliza `__repr__()` como alternativa.
- Um bom `__repr__()` deve ser claro, não ambíguo e, sempre que possível, representar um código capaz de recriar o objeto.

Esses métodos tornam nossas classes mais integradas ao Python e facilitam a depuração, a leitura do código e o desenvolvimento de bibliotecas profissionais.

## 📖 1.3.3 — O valor booleano de um tipo customizado

#### 🎯 Objetivo

Nesta seção aprenderemos como o Python determina se um objeto deve ser considerado **verdadeiro (`True`)** ou **falso (`False`)**.

Por meio dos métodos especiais relacionados ao valor booleano, podemos fazer com que nossas próprias classes se comportem naturalmente em estruturas como `if`, `while` e operadores lógicos, assim como acontece com listas, dicionários e outros tipos nativos do Python.

Esse comportamento torna o código mais intuitivo, expressivo e consistente com a linguagem, além de facilitar o desenvolvimento de bibliotecas e aplicações em Data Science, Machine Learning e Engenharia de Software.

```python
lista = []

if lista:
    print("Possui elementos")
else:
    print("Lista vazia")
```

```python
lista = [10, 20, 30]

if lista:
    print("Possui elementos")
else:
    print("Lista vazia")
```

Nota: O Python considera uma lista vazia como um valor False.

```python
if False:
    print("Possui elementos")
else:
    print("Lista vazia")
```

🧠 Regra para Memorizar

Objetos podem ser avaliados diretamente em um if. O Python decide automaticamente se eles representam True ou False.

```python
lista = [1]

if lista:
    print("True")
else:
    print("False")
```

"Este objeto possui um valor booleano verdadeiro?"

🧠 Regra para Memorizar

*  Coleções vazias → False
*  Coleções com elementos → True

```python
class Produto:
    pass

p = Produto()

if p:
    print("True")
else:
    print("False")
```

Porque, **por padrão**, todo objeto criado a partir de uma classe é considerado      
    **verdadeiro (True)**, a menos que a própria classe diga o contrário.

🧠 Regra para Memorizar

Objetos personalizados são True por padrão.

#### 💡 O mais importante

Perceba a evolução dos conceitos aprendidos até aqui.

O Python não executa diretamente funções e operadores sobre os objetos. Em vez disso, ele faz perguntas aos próprios objetos por meio dos **métodos especiais (dunder methods)**.

| Código | O que o Python pergunta ao objeto? | Método Especial |
|--------|-------------------------------------|-----------------|
| `len(obj)` | Qual é o seu tamanho? | `__len__()` |
| `abs(obj)` | Qual é o seu valor absoluto? | `__abs__()` |
| `obj1 + obj2` | Como devo realizar a soma? | `__add__()` |
| `print(obj)` | Como devo exibir este objeto para o usuário? | `__str__()` |
| `repr(obj)` | Como devo representar este objeto para o desenvolvedor? | `__repr__()` |
| `if obj:` | Você deve ser considerado verdadeiro ou falso? | *(Descobriremos nesta seção.)* |

📌 A grande ideia do Modelo de Dados do Python é que **o comportamento de um objeto é definido pelos métodos especiais implementados na sua classe**.

É isso que permite que objetos personalizados se comportem como tipos nativos da linguagem, tornando o código mais intuitivo, reutilizável e integrado ao ecossistema Python.
